```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
'''
Функция, которая посредством решения матричного уравнения
вычисляет коэффициенты естественного кубического сплайна.
'''
def qubic_spline_coeff(x_nodes, y_nodes):
    h = x_nodes[1:] - x_nodes[:-1]    # от h_1 до h_n-1
   
    left_diag = np.concatenate((h[:-1], [0]))    # левая диагональ
    general_diag = np.concatenate(([1], (h[1:] + h[:-1]) * 2, [1]))    # главная диагональ
    right_diag = np.concatenate(([0], h[1:]))    # правая диагональ
   
    a = y_nodes    # список коэффициентов а
   
    matr_A = np.diag(left_diag, -1) + np.diag(general_diag) + np.diag(right_diag, 1)    # трёхдиагональная матрица А
    matr_B = np.concatenate(([0], 3 * \
         (np.divide(np.subtract(a[2:], a[1:-1]), h[1:]) - \
         np.divide(np.subtract(a[1:-1], a[:-2]), h[:-1])), [0]))    # матрица В
    matr_A_inv = np.linalg.inv(matr_A)    # обратная матрица А
   
    c = np.array(matr_A_inv.dot(matr_B))    # список коэффициентов с
   
    b = np.subtract(np.divide(np.subtract(a[1:], a[:-1]), h), (c[1:] + 2 * c[:-1]) * h / 3)     # список коеффицинтов b
    d = np.array((c[1:] - c[:-1]) / h / 3)    # список коэффициентов d
    matrix = []     # матрица коэффициентов [ [b_1, c_1, d_1], ..., [b_n-1, c_n-1, d_n-1] ]
    n = len(x_nodes)
    for i in range(n - 1):
        matrix.append([b[i], c[i], d[i]])
    return matrix


'''
Функция, которая возвращает индекс x в списке узлов x_nodes
'''
def get_index(x, x_nodes):
    n = len(x_nodes)
    for i in range(n - 1):    
        if (x >= x_nodes[i] and x <= x_nodes[i + 1]):
            index = i
            break
    if (x > x_nodes[n - 1]):
        index = n - 2
    if (x < x_nodes[0]):
        index = 0
       
    return index

'''
Функция, которая возвращает значение кубического сплайна
в точке x (qs_coeff обозначает матрицу коэффициентов)
'''
def qubic_spline(x, qs_coeff, x_nodes, y_nodes):
    i = get_index(x, x_nodes)
    a = y_nodes[i]
    b = qs_coeff[i][0]
    c = qs_coeff[i][1]
    d = qs_coeff[i][2]
   
    return a + b * (x - x_nodes[i]) + c * (x - x_nodes[i])**2 + d * (x - x_nodes[i])**3

'''
Функция, которая возвращает значение производной кубического
сплайна в точке x (qs_coeff обозначает матрицу коэффициентов)
'''
def d_qubic_spline(x, qs_coeff, x_nodes, y_nodes):
    i = get_index(x, x_nodes)
    b = qs_coeff[i][0]
    c = qs_coeff[i][1]
    d = qs_coeff[i][2]
 
    return b + 2 * c * (x - x_nodes[i]) + 3 * d * (x - x_nodes[i])**2

'''
Функция, которая возвращает значение i-го базисного
полинома Лагранжа, заданного на узлах с абсциссами x_nodes,
в точке x.
'''
def l_i(i, x, x_nodes):
    n = len(x_nodes)
    polynom = 1
 
    for j in range(n):
        if i != j:
            polynom *= (x - x_nodes[j]) / (x_nodes[i] - x_nodes[j])
 
    return polynom


'''
Функция, которая возвращает значение интерполяционного
полинома Лагранжа, заданного на узлах с абсциссами x_nodes
и ординатами y_nodes, в точке x.
'''
def L(x, x_nodes, y_nodes):
    n = len(x_nodes)
    L_x = 0
   
    for i in range(n):
        L_x += y_nodes[i] * l_i(i, x, x_nodes)
 
    return L_x
```

```python
'''
Данные
'''
gdp_rus = [ 506500154001.466, 516814258695.568, 517962962962.963, 460290556900.726,
    435083713850.837, 395077301248.464, 395531066563.296, 391719993756.828,
    404926534140.017, 270953116950.026, 195905767668.562, 259708496267.33,
    306602673980.117, 345110438692.185, 430347770731.787, 591016690742.798,
    764017107992.391, 989930542278.695, 1299705247685.76, 1660844408499.61,
    1222643696991.85, 1524916112078.87, 2051661732059.78, 2210256976945.38,
    2297128039058.21, 2063662665171.89, 1365864126832.81, 1283162985989.3 ]
years = np.arange(1989., 2017., 1.)
left_board = 1989.
right_board = 2016.
left_board_extr = 1989.
right_board_extr = 2018.
x_nodes = years
y_nodes = gdp_rus
xlabels = ['год', 'год', 'год']
ylabels = ['трлн. $', 'трлн. $ / год', 'трлн. $']
 
'''
Рассчёт
'''
qs_coeff = qubic_spline_coeff(x_nodes, y_nodes)
x = np.linspace(left_board, right_board, 100)
S_x = [qubic_spline(x_i, qs_coeff, years, gdp_rus) for x_i in x]
d_S_x = [d_qubic_spline(x_i, qs_coeff, years, gdp_rus) for x_i in x]
L_x = [L(x_i, x_nodes, y_nodes) for x_i in x]
x_extr = np.linspace(left_board_extr, right_board_extr, 100)
S_x_extr = [qubic_spline(x_i, qs_coeff, years, gdp_rus) for x_i in x_extr]
d_S_x_extr = [d_qubic_spline(x_i, qs_coeff, years, gdp_rus) for x_i in x_extr]
```

```python
'''
Вывод
'''
fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(10, 20), dpi=380)
axes[0].plot(x, S_x, color="purple", linewidth=1.5, linestyle="-", label="S(x)")
axes[0].plot(x_nodes, y_nodes, 'o', color="red")
axes[0].set_xlabel(xlabels[0])
axes[0].set_ylabel(ylabels[0])
axes[0].legend(loc='upper left')
axes[0].grid()
axes[1].plot(x, d_S_x, color="blue", linewidth=1.5, linestyle="-", label="S'(x)")
axes[1].set_xlabel(xlabels[1])
axes[1].set_ylabel(ylabels[1])
axes[1].legend(loc='upper left')
axes[1].grid()
axes[2].plot(x_extr, S_x_extr, color="purple", linewidth=1.5, linestyle="-", label="S(x)")
axes[2].plot(x_nodes, y_nodes, 'o', color="red")
axes[2].set_xlabel(xlabels[0])
axes[2].set_ylabel(ylabels[0])
axes[2].legend(loc='upper left')
axes[2].grid()
axes[3].plot(x_extr, d_S_x_extr, color="blue", linewidth=1.5, linestyle="-", label="S'(x)")
axes[3].set_xlabel(xlabels[1])
axes[3].set_ylabel(ylabels[1])
axes[3].legend(loc='upper left')
axes[3].grid()
axes[4].plot(x, S_x, color="purple", linewidth=1.5, linestyle="-", label="S(x)")
axes[4].plot(x_nodes, y_nodes, 'o', color="red")
axes[4].plot(x, L_x, color="orange", linewidth=1.5, linestyle="-", label="L(x)")
axes[4].set_xlabel(xlabels[2])
axes[4].set_ylabel(ylabels[2])
axes[4].legend(loc='upper left')
axes[4].grid()
fig.tight_layout()
plt.show()
#fig.savefig("сubic_spline_graphics.png")
```
