```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from numpy import linspace
from math import cos, pi
```

```python
"""Функция, которая возвращает значение интерполируемой функции
в точке x
"""
def f(x):
    return 1 / (1 + 25 * x**2)
 
"""Функция, которая возвращает значение i-го базисного
полинома Лагранжа, заданного на узлах с абсциссами x_nodes,
в точке x
"""
def l_i(i, x, x_nodes):
    n = len(x_nodes)
    polynom = 1
 
    for j in range(n):
        if i != j:
            polynom *= (x - x_nodes[j]) / (x_nodes[i] - x_nodes[j])
 
    return polynom


"""Функция, которая возвращает значение интерполяционного
полинома Лагранжа, заданного наузлах с абсциссами x_nodes
и ординатами y_nodes, в точке x.
"""
def L(x, x_nodes, y_nodes):
    n = len(x_nodes)
    L_x = 0
   
    for i in range(n):
        L_x += y_nodes[i] * l_i(i, x, x_nodes)
 
    return L_x

"""Функция, которая возвращает вектор Чебышевских узлов
при их заданном количестве.
"""
def chebyshev_nodes(a, b, n):
    x_nodes = []
    for i in range(1, n + 1):
        x_nodes.append(0.5 * ((b - a) * cos((2 * i - 1) * pi / (2 * n)) + a + b))
 
    return x_nodes
```

```python
 
n = int(input("Please, write count of nodes: ")) # Колчество узлов
left_board = -1    # Левая граница диапазона узлов
right_board = 1   # Правая граница диапазона узлов
 
x_nodes = linspace(left_board, right_board, n)    # Вектор распределенных узлов
x_cheb_nodes = chebyshev_nodes(left_board, right_board, n)    # Вектор Чебышевских узлов
 
y_nodes = []    # Вектор аргументов для распределенных узлов
for i in range(n):
    y_nodes.append(f(x_nodes[i]))
    
y_cheb_nodes = []    # Вектор аргументов для Чебышевских узлов
for i in range(n):
    y_cheb_nodes.append(f(x_cheb_nodes[i]))
```

```python
x = np.linspace(left_board, right_board, 100, endpoint=True)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6), dpi=380)
axes[0].plot(x, f(x), color="blue", label="f(x)")    # Вывод исходной функции
axes[0].plot(x, L(x, x_nodes, y_nodes), color="orange", label="L(x)")    # Вывод интерполяционного полинома Лагранжа, заданного на распределенных узлах
axes[0].plot(x_nodes, y_nodes, 'o', color="red")
axes[0].legend(loc='upper left');
axes[0].grid()
axes[1].plot(x, f(x), color="blue", label="f(x)")    # Вывод исходной функции
axes[1].plot(x, L(x, x_cheb_nodes, y_cheb_nodes), color="purple", label="L(x_cheb)")    # Вывод интерполяционного полинома Лагранжа, заданного на Чебышевских узлах
axes[1].plot(x_cheb_nodes, y_cheb_nodes, 'o', color="red")
axes[1].legend(loc='upper left');
axes[1].grid()
fig.tight_layout()
plt.show()
# fig.savefig("lagrange_graphics_20_nodes.png")
```
